```python
from bs4 import BeautifulSoup
```

```python
import urllib.request
```

```python
response = urllib.request.urlopen('https://unal.edu.co/')
```

```python
html = response.read()
html
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
texto_plano = soup.get_text(strip=True)


with open('/content/url_txt.txt', 'w') as f:
    f.write(texto_plano)
```
